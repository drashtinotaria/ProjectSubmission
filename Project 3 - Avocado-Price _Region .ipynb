# Data Analysis - Linear Regression Model ==> Average Price

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error , mean_squared_error, r2_score,auc, roc_curve
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
lreg = LinearRegression()
knr = KNeighborsRegressor()
ls = Lasso()
en = ElasticNet()
dtr = DecisionTreeRegressor()
```

```python
# Using x as we getting 100 r2score

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 79 , test_size = 0.20)

x_train.shape ,x_test.shape,y_train.shape,y_test.shape
```

```python
#Best parameter for Linear Regression

parameters = { 'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
gs = GridSearchCV(lreg , parameters )
gs.fit(x_train,y_train)
gs.best_params_
```

```python
#Best parameter for KNearestNeighbors

parameters = { 'n_neighbors' : (1,10)}
gs = GridSearchCV(knr , parameters )
gs.fit(x_train,y_train)
gs.best_params_
```

```python
#Best parameter for Decision Tree

parameters = { "criterion": ["mse", "mae"] }
gs = GridSearchCV(dtr , parameters )
gs.fit(x_train,y_train)
gs.best_params_
```

```python
lreg = LinearRegression(copy_X = 'True' , fit_intercept = 'True' , normalize = 'False')
knr = KNeighborsRegressor(n_neighbors= 10)
dtr = DecisionTreeRegressor(criterion = 'mse')
svr = SVR(C=1, kernel='rbf')
```

```python
# Best Parameter for Random Forest

rf = RandomForestRegressor()
parameters = {  "n_estimators" : [10 , 100 , 500 , 1000]}
gs = GridSearchCV(rf , parameters , cv =5)
gs.fit(x_train,y_train)
gs.best_params_
```

```python
# Best Parameter for Gradient Boosting

gbr = GradientBoostingRegressor()
parameters = { "learning_rate" : [0.001 , 0.01 , 0.1 , 1] , "n_estimators" : [10 , 100 , 500 , 1000]}
gs = GridSearchCV(gbr , parameters , cv =5)
gs.fit(x_train,y_train)
gs.best_params_
```

```python
gbr = GradientBoostingRegressor(learning_rate = 0.1 ,n_estimators = 1000 )
rf = RandomForestRegressor(n_estimators = 1000)
```

```python
knr.fit(x_train,y_train)
pre = knr.predict(x_test)
r2 = r2_score(y_test,pre)
rme = np.sqrt(mean_squared_error(y_test,pre))

print(r2 , "\n" , rme)
```

```python
models = []
models.append(('KNeighborsRegressor' , knr))
models.append(('LinearRegression' , lreg))
models.append(('Lasso' , ls))
models.append(('ElasticNet' , en))
models.append(('GradientBooster' , gbr))
models.append(('RandomForestRegressor' , rf))
models.append(('SVR' , svr))

models
```

```python
Model = []
r2score = []
cv = []
rmse = []

for name,model in models:
    print("***************************" , name , "***************************\n")
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pred = model.predict(x_test)
    print("\n")
    r2s = r2_score(y_test,pred)
    print("r2 score :" , r2s)
    r2score.append(r2s*100)
    print("\n")
    sc = cross_val_score(model , x, y , cv = 10 , scoring = 'r2')
    print("Cross_Val_score : " , sc)
    cv.append(sc*100)
    print("\n")
    rme = np.sqrt(mean_squared_error(y_test,pred))
    print("RMSE : " , rme)
    rmse.append(rme)
    print("\n\n")
    
```

```python
result = pd.DataFrame({'Model' : Model , 'R2Score' : r2score , "RMSE" : rmse , "Cross_Val_Score" :cv})

result
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report, auc, roc_curve
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
lreg = LogisticRegression()
knr = KNeighborsClassifier()
dtc = DecisionTreeClassifier()
svc = SVC()
```

```python
#Best parameter for Logistic Regression

from sklearn.model_selection import GridSearchCV

parameters = { 'penalty' : ('11', '12' , 'none')}
gs = GridSearchCV(lreg , parameters )
gs.fit(x_train,y_train)
gs.best_params_
```

```python
#Best parameter for SVC

parameters = { 'kernel' : ('linear' , 'rbf') , 'C' : [1,10]}
gs = GridSearchCV(svc , parameters )
gs.fit(x_train,y_train)
gs.best_params_
```

```python
lreg.fit(x_train,y_train)
pred = lreg.predict(x_test)
accuracyscore = accuracy_score(y_test,pred)
confmatrix = confusion_matrix(y_test,pred)
classreport = classification_report(y_test,pred)
print(accuracyscore,"\n")
print(confmatrix)
```
